```python
missing_gdp = df[df['GDP Growth'].isnull()]
missing_gdp_by_region = missing_gdp['region'].value_counts()
print(missing_gdp_by_region)
```

```python
df['GDP Growth'] = df.groupby('region')['GDP Growth'].transform(lambda x: x.fillna(x.mean()))
print(df['GDP Growth'].isnull().sum())
```

```python
df['borders'] = df['borders'].fillna(0)
print(df['borders'].isnull().sum())
```

```python
df['Interest Rate'] = df.groupby('region')['Interest Rate'].transform(lambda x: x.fillna(x.mean()))
print(df['Interest Rate'].isnull().sum())
```
